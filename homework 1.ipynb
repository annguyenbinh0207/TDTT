### bai6

```python
import math

def tinh_dien_tich_tam_giac():
    a = int(input())
    b = int(input())
    c = int(input())
    if a + b > c and a + c > b and b + c > a:
        s = (a + b + c) / 2
        dien_tich = math.sqrt(s * (s - a) * (s - b) * (s - c))
        return f"{dien_tich:.1f}" 
    else:
        return "ko phai 3 canh cua tam giac" 
print(tinh_dien_tich_tam_giac())
```

### bai8

```python
ten = input("Ten chu ho: ")
truoc = int(input("Chi so thang truoc: "))
nay = int(input("Chi so thang nay: "))

kwh = nay - truoc
tien = 0

if kwh <= 50:
    tien = kwh * 1984
else:
    # Nếu dùng > 50 số, máy nhảy vào đây
    if kwh <= 100:
        tien = 50 * 1984 + (kwh - 50) * 2050
    else:
        # Nếu dùng > 100 số, máy lại nhảy vào đây
        if kwh <= 200:
            tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
        else:
            # Nếu dùng > 200 số...
            if kwh <= 300:
                tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
            else:
                # Nếu dùng > 300 số...
                if kwh <= 400:
                    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
                else:
                    # Bậc cuối cùng: > 400 số
                    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

# Tính VAT 8% và in kết quả
tong_tien = int(tien * 1.08)
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {tong_tien}")
```
