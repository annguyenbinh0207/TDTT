```python
a = float(input())
b = float(input())
m = float(input())
c = float(input())
d = float(input())
n = float(input())

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    print(Dx / D, Dy / D)
else:
    if Dx == 0 and Dy == 0:
        print("VSN")
    else:
        print("VN")
```
