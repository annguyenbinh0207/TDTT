### W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
for k in range(n):
    if 2**k == n:
        print(" n co dang 2^k")
    if 2**k > n:
        break
```

### W4A23
### Nhập n
### Cho S(k) = 1 + 2 + 3 + … + k
### Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
S = 0
k = 1
while S + k < n:
    S += k
    k += 1
print(S)
```

### W4A24 Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 



```python
A = int(input())
S = 0
n = 0
while S <= A:
    n +=1
    S +=1/n
print(n)
```

### W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
numbers = []
while True:
    n = int(input())
    if n == -1:
        break
    numbers.append(n)
if len(numbers) > 0:
    print(max(numbers))
    print(min(numbers))
else:
    print("Ban chua nhap so nao hop le")
```

### W4A26 Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
a = int(input())
f1 = 1
f2 = 1
for i in range(a):
   f_sau = f1 + f2
   if f_sau > a:
        break
   f1 = f2
   f2 = f_sau
print(f2)
```

### W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. 

```python
chuoi= input()
ketqua =""
for i in chuoi:
    if i.isupper:
        ketqua += i.lower()
    elif i.islower:
        ketqua += i.upper()
    else:
        ketqua = i
print(ketqua)
```

### W4A31 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
chuoi = input() 
hoa = 0
thuong = 0 
so = 0
for i in chuoi:
    if i.isupper():
        hoa +=1
    elif i.islower():
        thuong +=1
    else:
        so += 1
print("hoa =", hoa,"thuong =", thuong,"so =" , so)
```
