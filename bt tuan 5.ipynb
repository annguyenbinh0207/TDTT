### W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
n = int(input())
if prime(n):
    print("n la so nguyen to")
else:
    print("n ko la so nguyen to")    
```

### W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def perfect(n):
    if n <=1:
        return False
    uoc = 0
    for i in range(1,n//2+1):
        if n % i == 0:
            uoc += i
    return uoc == n
n = int(input())

if perfect(n):
    print(n, "la so hoan hao")
else:
    print(n, "ko la so hoan hao")
    
```

### W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def banhmiphap(list, k):
    for i in range(len(list)):
        if list[i] == k:
            return i 
    return -1 
list = [10, 20, 30, 40, 20]
k = 20
vi_tri = banhmiphap(list, k)
print(vi_tri+1)
```

### W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    uoc = 1
    for i in range(1,n+1):
        uoc *= i
    return uoc
n = int(input())
ketqua = giai_thua(n)
print(ketqua)
```

 Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat\)
 Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
 Phép trừ \( - \)
 Phép cộng \( + \)

 Kết quả làm tròn đến chữ số thập phân thứ hai.
 Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
 Đầu ra: Kết quả phép toán.

##### W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn

```python
def hamming_distance(x, y):
    return (x ^ y).bit_count()
print(hamming_distance(1, 3))
```

### W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False

```python
def divisible(a, b):
    tich = a*b
    if tich % 10==0:
        return True
    else:
        return False
a, b = map(int,input().split())
print(divisible(a, b))
```

### W5A12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|. 


```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    return(q, r)
kq = divide_with_remainder(7, 2)
print(kq)
```

### W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def tamgiacvuong(a, b, c):
    if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == b**2 + a**2:
        return True
    else:
        False
a,b,c = map(int,input().split())
print(tamgiacvuong(a, b, c))
```
